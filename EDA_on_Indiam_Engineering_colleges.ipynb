```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as bo
import io
#import warning to ignore any warning
import warnings
warnings.filterwarnings('ignore')
```

##Importing CSV file

```
#importing CSV file for indian engineering colleges dataset
from google.colab import files


uploaded = files.upload()

```

```
df = pd.read_csv(io.BytesIO(uploaded['Indian_Engineering_Colleges_Dataset.csv']))
print(df)
```

#Permorming basic *EDA*

-Removing sr.no column as it is of no use

```
df.drop('Sr.no',axis=1,inplace=True)
```

```
df.tail()
```

#Performing advance EDA
-Converting datatype from object to float to perform numerical operation on the attribute

```

df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Academic'] = pd.to_numeric(df['Academic'], errors='coerce')
df['Faculty'] = pd.to_numeric(df['Faculty'], errors='coerce')
df['Infrastructure'] = pd.to_numeric(df['Infrastructure'], errors='coerce')
df['Placement'] = pd.to_numeric(df['Placement'], errors='coerce')
df['Social_Life'] = pd.to_numeric(df['Social_Life'], errors='coerce')
df['Accommodation'] = pd.to_numeric(df['Accommodation'], errors='coerce')
```

## Rating Statistics

```
# Group the data by the state and calculate the average rating for each state
average_rating_by_state = df.groupby('State')['Rating'].mean().reset_index()

# Sort the results by average rating in descending order
average_rating_by_state = average_rating_by_state.sort_values(by='Rating', ascending=False)

# Create a bar chart for ratings
plt.figure(figsize=(10, 6))
plt.bar(average_rating_by_state['State'], average_rating_by_state['Rating'], color='orange')

# Adding labels and title
plt.xlabel('State')
plt.ylabel('Average Rating')
plt.title('Average Rating by State')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```

##Placement Static

```
plt.figure(figsize=(10, 6))
bo.boxplot(x='Placement', data=df)
plt.xlabel('Placement')
plt.title('Placement Statistics')
plt.show()
```

```
# Group the data by the state and calculate the total placements for each state
total_placements_by_state = df.groupby('State')['Placement'].sum()

# Create a pie chart for placements
plt.figure(figsize=(8, 8))
plt.pie(total_placements_by_state, labels=total_placements_by_state.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)

# Adding title
plt.title('Placement Distribution by State')

# Display the plot
plt.axis('equal')
plt.tight_layout()
plt.show()
```

**State wise placement rating**

-Maharashtra have the highest no of Placement

```
# Creating a histogram
plt.figure(figsize=(8, 6))
bo.histplot(data=df, x='Faculty', bins=10, kde=True)
plt.xlabel('Faculty Ratings')
plt.ylabel('Frequency')
plt.title('Distribution of Faculty Ratings')
plt.show()
```

- From the above Graph it is clear that most of the faculty in indian colleges are good having a rating of 8 out of 10

```
# Create a bar plot for the Rating attribute
plt.figure(figsize=(8, 6))
rating_counts = df['Rating'].value_counts().sort_index()
rating_counts.plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings in Indian Engineering Colleges')
plt.show()
```

-From the above graph it is clear that most of the colleges in india has given a rating of 7.7 which is above average

```
# Group the data by the state and calculate the average placement for each state
average_placement_by_state = df.groupby('State')['Placement'].mean().reset_index()

# Sort the results by average placement in descending order to see the highest placement states first
average_placement_by_state = average_placement_by_state.sort_values(by='Placement', ascending=False)

# Display the results
print(average_placement_by_state)
```

- Thus we can conclude that college in Maharshtra have the highest number of placements followed by tamil nadu

```
# Group the data by the state and calculate the average social life for each state
average_social_life_by_state = df.groupby('State')['Social_Life'].mean().reset_index()

# Sort the results by average social life in descending order to see the highest social life states first
average_social_life_by_state = average_social_life_by_state.sort_values(by='Social_Life', ascending=False)

# Display the results
print(average_social_life_by_state)
```

- Hence we can see that the social life in Maharasthra college is also rated the highest

-Infrastucture rating of all the colleges state-wise

```
# Group the data by the state and calculate the average infrastructure for each state
average_infrastructure_by_state = df.groupby('State')['Infrastructure'].mean().reset_index()

# Sort the results by average infrastructure in descending order
average_infrastructure_by_state = average_infrastructure_by_state.sort_values(by='Infrastructure', ascending=False)

# Create a bar chart for infrastructure
plt.figure(figsize=(10, 6))
plt.bar(average_infrastructure_by_state['State'], average_infrastructure_by_state['Infrastructure'], color='teal')

# Adding labels and title
plt.xlabel('State')
plt.ylabel('Average Infrastructure')
plt.title('Average Infrastructure by State')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```

-Comparing thr Accomodation of all the colleges state wise

```
# Group the data by the state and calculate the average accommodation for each state
average_accommodation_by_state = df.groupby('State')['Accommodation'].mean().reset_index()

# Sort the results by average accommodation in descending order
average_accommodation_by_state = average_accommodation_by_state.sort_values(by='Accommodation', ascending=False)

# Create a bar chart for accommodation
plt.figure(figsize=(10, 6))
plt.bar(average_accommodation_by_state['State'], average_accommodation_by_state['Accommodation'], color='green')

# Adding labels and title
plt.xlabel('State')
plt.ylabel('Average Accommodation')
plt.title('Average Accommodation by State')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()




```

-Comparing the acedemic record of all the colleges state wise

```
# Grouping the data by the state and calculate the average academic performance for each state
average_academic_by_state = df.groupby('State')['Academic'].mean().reset_index()

# Sorting the results by average academic performance in descending order
average_academic_by_state = average_academic_by_state.sort_values(by='Academic', ascending=False)

# Create a bar chart for academic performance
plt.figure(figsize=(10, 6))
plt.bar(average_academic_by_state['State'], average_academic_by_state['Academic'], color='purple')

# Adding labels and title
plt.xlabel('State')
plt.ylabel('Average Academic Performance')
plt.title('Average Academic Performance by State')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```
